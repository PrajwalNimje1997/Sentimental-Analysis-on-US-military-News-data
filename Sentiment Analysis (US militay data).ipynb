# Text Cleaning

```python
shortword = re.compile(r"\b\w{1,2}\b ")
Url = re.compile(r"https?:*/+[a-zA-Z0-9./]*")
unwanted = ' `~!@#$%^&*()_=+-{:}|[]\;'",./<>?"'1234567890 '
stop_words = set("/Users/prajwalite/Downloads/StopWords_GenericLong.txt")
def clean(text):
    #Escaping out HTML characters
    #text=HTMLParser().unescape(text)
    #library for regular expressions
    text = text.lower()
 
    # remove hyperlinks
    text = re.sub(r'https?:\/\/.\S+', "", text)
 
    # remove hashtags
    # only removing the hash # sign from the word
    text = re.sub(r'#', '', text)
    
    # remove old style retweet text "RT"
    text = re.sub(r'^RT[\s]+', '', text)
    text = re.sub(Url,'',text)
    text = re.sub(shortword,'',text)
    text = text.replace('@','')
    text = text.replace('#','')
    text = text.replace(':','')
    text = text.replace('>','')
    text = text.replace('<','')
    text = text.replace("/" , '')
    text = text.replace(" "" " , '')
    
    text = text.replace('`','')
    text = text.replace('~','')
    text = text.replace('!','')
    text = text.replace('$','')
    text = text.replace('%','')
    text = text.replace("^" , '')
    text = text.replace(" & " , '')
    
    text = text.replace(',','')
    text = text.replace('/','')
    text = text.replace("|" , '')
    text = text.replace('|','')
    text = text.replace('+','')##
    text = text.replace("," , '')
    text = text.replace(" === " , '')
    
    text = text.replace('*','')
    text = text.replace('-','')
    text = text.replace('.','')
    text = text.replace('(','')
    text = text.replace(')','')
    text = text.replace('_','')
    text = text.replace('[','')
    text = text.replace("]" , '')
    text = text.replace(" { " , '')
    text = text.replace("}" , '')
    
    text = text.replace("1" , '')
    text = text.replace('2','')
    text = text.replace('3','')
    text = text.replace('4','')
    text = text.replace('5','')
    text = text.replace('6','')
    text = text.replace('7','')
    text = text.replace('8','')
    text = text.replace("9" , '')
    text = text.replace("0" , '')
    
    text = text.split()
    text = [word for word in text if word not in stop_words]
    text = " ".join(text)
    
    return text
```

# Positive Negative score

```python
pnScore = pd.read_csv("/Users/prajwalite/Downloads/LoughranMcDonald_MasterDictionary_2020.csv")
#pnScore = pnScore[11:17]
pnScore
```

```python
pos = [] #positive words list from above csv file
neg = [] #negative words list from above csv file
for i in pnScore.index:
    if (pnScore['Negative'][i]) < (pnScore['Positive'][i]):
        pos.append(pnScore['Word'][i])
    elif (pnScore['Negative'][i]) > (pnScore['Positive'][i]):
        neg.append(pnScore['Word'][i])
    else:
        continue
```

```python
pos = [x.lower() for x in pos]
neg = [x.lower() for x in neg]
```

```python
posDF = []        #list-storing no. of pos words
negDF = []        #list-storing no. of neg words
posPropDF = []    #list-pos words/len(words)
negPropDF = []    #list-neg words/len(words)
polarityDF = []   #list-storing polarity score
subjectiveDF = [] #list-storing subjective score

def pnValue(words):
    posValue = 0
    negValue = 0
    for word in words:
        if word in pos:
            posValue += 1
        elif word in neg:
            negValue += 1
        else:
            continue
    pos_proportion = posValue/len(words)
    neg_proportion = negValue/len(words)
    posDF.append(posValue)
    negDF.append(negValue)
    posPropDF.append(pos_proportion)
    negPropDF.append(neg_proportion)
    #Polarity score
    polValue = (posValue-negValue)/((posValue+negValue)+0.000001)
    polarityDF.append(polValue)
    #Subjective Score
    subValue = (posValue+negValue)/((len(words))+0.000001)
    subjectiveDF.append(subValue)
    
```

# Average Number of Words Per Sentence

```python
wPerS_DF = [] #list-storing- Average Number of Words Per Sentence
def wPerS(words,text):
    Sentences = sent_tokenize(text)
    wPerS = len(words)/len(Sentences)
    wPerS_DF.append(wPerS)
```

# Complex Words

```python
complexWordCount_DF = [] #no. of complex words
prop_complexDF = []      #prop. of complex words
complexDF = []           #percentage of complex words
def complexWords(words):
    complexWordCount = 0
    for oneWord in words:
        syllableCount=0
        for w in oneWord:
            if(w=='a' or w=='e' or w=='i' or w=='o' or w=='u' or w=='A' or w=='E' or w=='I' or w=='O' or w=='U'):
                syllableCount += 1
        if syllableCount >= 2:
            complexWordCount += 1
        else:
            continue
    complexWordCount_DF.append(complexWordCount)
    prop_ComplexWords = complexWordCount/len(words)
    percComplexWords = (complexWordCount/len(words)) * 100
    prop_complexDF.append(prop_ComplexWords)
    complexDF.append(percComplexWords)
```

# Fog index

```python
fogIndex_DF = [] #storing fog index value
def fogIndex(wPerS_DF,prop_complexDF):
    for i in range(len(wPerS_DF)):
        fog = 0.4*(wPerS_DF[i] + prop_complexDF[i])
        fogIndex_DF.append(fog)
```

# Uncertainity Score

```python
uncertainityWordsList = [] #list-storing uncertainity words from below given csv file
uncertainityCSV = pd.read_csv("/Users/prajwalite/Downloads/uncertainty_dictionary.xlsx - Sheet1.csv")
uncertainityCSV
for i in uncertainityCSV['Word']:
    uncertainityWordsList.append(i.lower())   
```

```python
uncertainityScore_DF = []   #storing uncertainity score
uncertainProp_DF = []       #storing uncertainity proportion
def uncertainityScore(words):
    uncertainValue = 0
    for i in words:
        if i in uncertainityWordsList:
            uncertainValue+=1
        else:
            continue
    uncertainProp = uncertainValue/len(words)   #find uncertainity proportion 
    uncertainityScore_DF.append(uncertainValue)
    uncertainProp_DF.append(uncertainProp)
```

# Constraining Score

```python
constrainingWordsList = []   #list-storing constraining words from below given csv file
constrainingCSV = pd.read_csv("/Users/prajwalite/Downloads/constraining_dictionary.xlsx - Sheet1.csv")
constrainingCSV
for i in constrainingCSV['Word']:
    constrainingWordsList.append(i.lower())
```

```python
constrainingScore_DF = []    #storing constraining score
constrainingProp_DF = []     #storing constraining proportion

def constrainingScore(words):
    constrainingValue = 0
    for i in words:
        if i in constrainingWordsList:
            constrainingValue+=1
        else:
            continue
    constrainingProp = constrainingValue/len(words)  #find constraining proportion
    constrainingScore_DF.append(constrainingValue)
    constrainingProp_DF.append(constrainingProp)
```

# Constraining words whole report

```python
def constrainWholeReport():
    constrainWhole = 0
    for i in constrainingScore_DF:
        constrainWhole+=i
    constrainingWhole_DF = [constrainWhole]*152
    return constrainingWhole_DF
```

# Main

```python
length = []
wordCount_DF = []
for i in df['weblink']:
    url = i
    #get name of all filings 
    download = requests.get( url , headers={'User-Agent':'Brave'}).content
    text = download.decode("utf-8")#.split('\n')
    #text cleaning
    cleaned = clean(text)
    #tokenization
    words = word_tokenize(cleaned)
    wordCount = len(words)
    wordCount_DF.append(wordCount)
    #find positive and negative score
    pnValue(words)
    #find Average Number of Words Per Sentence
    wPerS(words,text)
    #find the percentage of complex words
    complexWords(words)
    #uncertainity Score
    uncertainityScore(words)
    #constraining Score
    constrainingScore(words)
```

```python
#Constrain score of whole report
constrainWhole_VALUE = constrainWholeReport()
```

```python
#Calculate fog index
fogIndex(wPerS_DF,prop_complexDF)
```

```python
df['positive_score'] = posDF
df['negative_score'] = negDF
df['polarity_score'] = polarityDF
df['average_sentence_length'] = wPerS_DF
df['percentage_of_complex_words'] = complexDF
df['fog_index'] = fogIndex_DF
df['complex_word_count'] = complexWordCount_DF
df['word_count'] = wordCount_DF
df['uncertainty_score'] = uncertainityScore_DF
df['constraining_score'] = constrainingScore_DF
df['positive_word_proportion'] = posPropDF
df['negative_word_proportion'] = negPropDF
df['uncertainty_word_proportion'] = uncertainProp_DF
df['uncertainty_word_proportion'] = uncertainProp_DF
df['constraining_word_proportion'] = constrainingProp_DF
df['constraining_words_whole_report'] = constrainWhole_VALUE
```
